# Define Custom Layer

```python
# so far it does not include padding
# tensor shape must be (batch,channel,x,y,z)


class SparseConv3D(nn.Module):
    
    def __init__(self, in_channels = 1, out_channels = 100, kernel_size = (3,3,3), stride = (2,2,2), padding = (0,0,0), bias = True, device='cuda'):
        
        super().__init__()
        self.out_channels = out_channels
        self.in_channels = in_channels
        self.kernel_size = kernel_size
        self.stride = stride
        self.padding = padding
        self.device = device

        
        # Initialize weights and biases
        self.weights = nn.Parameter( torch.randn( (out_channels, in_channels, kernel_size[0], kernel_size[1], kernel_size[2]) ) )
        if bias == True:
            self.bias = nn.Parameter(torch.randn(out_channels))
        else:
            self.bias = torch.zeros(out_channels)
                    
    def forward(self, x):
        
        # coalesce the sparse tensor to make sure data can be extraxted
        x = x.coalesce()
    
        # Extract info from sparse tensor
        indices = x.indices()
        values = x.values()
        shape = x.shape
        out_shape = (shape[0], self.out_channels, int( ( shape[2] + 2*self.padding[0] - self.kernel_size[0] ) / self.stride[0] + 1 ), int( ( shape[3] + 2*self.padding[1] - self.kernel_size[1] ) / self.stride[1] + 1 ), int( ( shape[4] + 2*self.padding[2] - self.kernel_size[2] ) / self.stride[2] + 1 ) ) 
        
        # Store channel for each entry
        channel = indices[1]

        # Create Features - a tensor of all channel values for a given batch / x / y / z position 
        Features = []
        for ind in range(self.in_channels):
            Features.append( values[channel == ind] )           
        Features = torch.stack(Features)
        
        # Remove duplicate batch / x / y / z position  info due to multiple channels
        batch = indices[0][channel == 0]
        x = indices[2][channel == 0]
        y = indices[3][channel == 0]
        z = indices[4][channel == 0]
        
        # for x/y/z compute (index - all kernel indices along same dimension) / stride
        a = ( x.repeat(self.kernel_size[0],1) -  torch.arange(self.kernel_size[0])[:, None].to(self.device) ) / self.stride[0]
        b = ( y.repeat(self.kernel_size[1],1) -  torch.arange(self.kernel_size[1])[:, None].to(self.device) ) / self.stride[1]
        c = ( z.repeat(self.kernel_size[2],1) -  torch.arange(self.kernel_size[2])[:, None].to(self.device) ) / self.stride[2]
        
        # Compute corresponding truth values (True if an element is an integer, false otherwise)
        # True if it can go into the output shape
        Ta = (-2.0*(torch.fmod(a,1)-0.5)).type(torch.BoolTensor).to(self.device) * (a < out_shape[2])  * (a >= 0)
        Tb = (-2.0*(torch.fmod(b,1)-0.5)).type(torch.BoolTensor).to(self.device) * (b < out_shape[3])  * (b >= 0)
        Tc = (-2.0*(torch.fmod(c,1)-0.5)).type(torch.BoolTensor).to(self.device) * (c < out_shape[4])  * (c >= 0)
        
        out_sparse_tensor = torch.sparse_coo_tensor(out_shape).to(self.device)
        
        for f in torch.arange(self.out_channels):
            for i in torch.arange(self.kernel_size[0]):
                for j in torch.arange(self.kernel_size[1]):
                    for k in torch.arange(self.kernel_size[2]):

                        T=Ta[i]*Tb[j]*Tc[k]
                        
                        out_ind_0 = batch[T]
                        out_ind_1 = torch.ones(len(out_ind_0)).to(self.device) * f #indices indicating which filter is used
                        out_ind_x = a[i,T]
                        out_ind_y = b[j,T]
                        out_ind_z = c[k,T]

                        out_vals = torch.sum( Features[:,T] * ( self.weights[f,:,i,j,k].reshape( self.in_channels, 1 ) ), 0 )
                        
                        inds = torch.stack([out_ind_0,out_ind_1,out_ind_x,out_ind_y,out_ind_z])

                        out_sparse_tensor += torch.sparse_coo_tensor(inds,out_vals,out_shape)

            out_sparse_tensor #add bias here
        
        
        return out_sparse_tensor
```

# Define Convnet Architecture

```python
# Get cpu or gpu device for training.
device = "cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu"
print(f"Using {device} device")


# Define model
class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv3d(in_channels=1, out_channels=50, kernel_size=6, stride=2)
        self.pool = nn.MaxPool3d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv3d(in_channels=50, out_channels=30, kernel_size=4, stride=1)
        self.conv3 = nn.Conv3d(in_channels=30, out_channels=20, kernel_size=3, stride=1)
        self.fc1 = nn.Linear(12**3 *20, 100)
        self.fc2_1 = nn.Linear(100, 30)
        self.fc3_1 = nn.Linear(30, 3)
        self.fc2_2 = nn.Linear(100, 30)
        self.fc3_2 = nn.Linear(30, 1)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = F.relu(self.conv2(x))
        x = self.pool(F.relu(self.conv3(x)))
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x1 = torch.tanh(self.fc2_1(x))
        x1 = F.normalize(self.fc3_1(x1),dim=1)
        x2 = F.relu(self.fc2_2(x))
        x2 = F.softplus(self.fc3_2(x2))
        return x1,x2

model = NeuralNetwork().to(device)

# Print a nicer model summary
print(model)
```
